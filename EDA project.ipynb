# Data Cleaning, Data Preparation and Feature Engineering

```python
import pandas as pd
import numpy as np

# 'np.nan' creates a missing value
data = {
    'House_ID': [1, 2, 3, 4, 5, 6],
    'Price': [250000, 300000, np.nan, 400000, 150000, 280000],
    'SqFt': [1500, 1800, 2200, 2500, 1000, np.nan],
    'City': ['Austin', 'Dallas', 'Austin', 'Houston', 'Dallas', 'Austin'],
    'Condition': ['Good', 'Excellent', 'Fair', 'Excellent', np.nan, 'Good']
}

df = pd.DataFrame(data)
print("--- Raw Data ---")
df
```

```python
df['Price'] = df['Price'].fillna(df['Price'].median())
```

```python
df
```

```python
df['SqFt'] = df['SqFt'].fillna(df['SqFt'].median())
```

```python
df
```

```python
df['Condition'].mode()[0]
```

```python
df['Condition'] = df['Condition'].fillna(df['Condition'].mode()[0])
```

```python
df
```

# Encoding

```python
df_encoded = pd.get_dummies(df, columns=['City'])
```

```python
df_encoded
```

```python
# Feature Engineering -> Create new features from exisiting features
df['Price_per_sqft'] = df['Price']/df['SqFt']
```

```python
df
```

```python
hr_data = {
    'Candidate': ['A', 'B', 'C', 'D', 'E'],
    'Test_Score': [85, 90, np.nan, 75, 88],
    'Experience_Yrs': [2, 5, 3, np.nan, 4],
    'Degree': ['Bachelor', 'Master', 'PhD', 'Bachelor', 'Master']
}
df_hr = pd.DataFrame(hr_data)
```

```python
# Tasks:

#1. Check for missing values.

#2. Fill missing Test_Score with the Mean.

#3. Fill missing Experience_Yrs with the Median.

#4. Perform Label Encoding on the Degree column manually using a map.

# Map: Bachelor=1, Master=2, PhD=3. (Since PhD > Master > Bachelor, this order matters!).

#5. Create a new feature called Score_Per_Year (Test_Score / Experience_Yrs).
```

```python
df_hr.isnull().sum()
```

```python
df_hr['Test_Score'] = df_hr['Test_Score'].fillna(df_hr['Test_Score'].mean())
```

```python
df_hr
```

```python
df_hr['Experience_Yrs'] = df_hr['Experience_Yrs'].fillna(df_hr['Experience_Yrs'].median())
```

```python
df_hr
```

```python
df_hr.isnull().sum()
```

```python
degree_map = {'Bachelor': 1, 'Master': 2, 'PhD': 3}
df_hr['Degree_encoded'] = df_hr['Degree'].map(degree_map)
```

```python
df_hr
```

```python
df_hr['Score_per_year'] = df_hr['Test_Score'] / df_hr['Experience_Yrs']
```

```python
df_hr
```

# Titanic Case Study

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
df = sns.load_dataset('titanic')
```

```python
df.head(5)
```

```python
df.info()
```

```python
# Univariate Analysis
# Q. How many survived vs how many died
# Countplot , barplot
sns.countplot(x='survived', data=df)
plt.xticks([0, 1], ['Died', 'Survived'])
```

More people died compared to people who survived, almost double.

```python
# Age distribution
sns.histplot(df['age'].dropna(), kde=True)
```

```python
sns.histplot(df['age'].fillna(df['age'].median()))
```

```python
# Fare Distribution
sns.histplot(df['fare'].dropna(), kde=True)
```

Observation:

Age: Roughly bell-shaped, peaking around 20-30 years old. Note the small bump near 0-5 (infants).

Fare: Highly Right-Skewed. Most tickets were cheap, but a few were incredibly expensive (Outliers).
